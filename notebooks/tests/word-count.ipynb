# test code to do word count of a corpus

use CountVectorizer to generate word counts for our corpus

```python
# import sibling utilities
import sys
sys.path.append('../..')

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
infile = "../../dataset/amazon_reviews/amazon_reviews_us_Wireless_v1_00-test-preprocessed.csv"
df = pd.read_csv(infile)
```

```python
corpus = df["review_body"]
cv = CountVectorizer()
cv_matrix = cv.fit_transform(corpus.array)

vocab = cv.get_feature_names()
print(f'vocab_length: {len(vocab)}')
df = pd.DataFrame(cv_matrix.toarray(), columns=vocab)

df
```

```python
wc_df = pd.DataFrame(df.sum()).reset_index().rename({0: "occurrence", "index": "word"}, axis=1).sort_values("occurrence", ascending=False)
wc_df
```
